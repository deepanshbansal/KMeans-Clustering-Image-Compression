```python
from PIL import Image
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img=Image.open('bird.png')
img
```

```python
img_np=np.asarray(img)
img_np.shape
```

```python
pixels=img_np.reshape((img_np.shape[0]*img_np.shape[1]),img_np.shape[2])
pixels.shape
```

```python
model=KMeans(n_clusters=6)
model.fit(pixels)
```

```python
pixels_centroids=model.labels_
pixels_centroids.shape
```

```python
cluster_centers=model.cluster_centers_
cluster_centers.shape
```

```python
final=np.zeros((pixels_centroids.shape[0],3))
final.shape
```

```python
for cluster_no in range(6):
    final[pixels_centroids==cluster_no]=cluster_centers[cluster_no]
print(final)
```

```python
comp_img=final.reshape(img_np.shape[0],img_np.shape[1],3)
comp_img.shape
```

```python
comp_img=Image.fromarray(np.uint8(comp_img))
comp_img.save('bird1.png')
```

```python
fname='bird.png'
fname1='bird1.png'
```

```python
img1=mpimg.imread(fname,0)
img2=mpimg.imread(fname1,0)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(20,20))
ax1.imshow(img1)
ax1.set_title('Original Image')
ax2.imshow(img2)
ax2.set_title('Compressed Image')
plt.show()
```

```python
import os
print('size of original image',int(os.stat('bird.png').st_size/1024),'Kb')
print('size of compressed image',int(os.stat('bird1.png').st_size/1024),'Kb')
```
